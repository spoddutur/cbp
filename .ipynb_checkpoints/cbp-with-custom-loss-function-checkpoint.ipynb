```python
def triplet_loss(y_true, y_pred, epsilon=1e-8):
    """
    Implementation of the triplet loss as defined by formula (3)
    
    Arguments:
    y_true -- true labels, required when you define a loss in Keras, you don't need it in this function.
    y_pred -- python list containing three objects:
            anchor -- the encodings for the anchor images, of shape (None, 128)
            positive -- the encodings for the positive images, of shape (None, 128)
            negative -- the encodings for the negative images, of shape (None, 128)
    
    Returns:
    loss -- real number, value of the loss
    """
    
    label_positive, label_negative = y_true[0], y_true[1]
    pred_positive, pred_negative = y_pred[0], y_pred[1]
    el = -(tf.multiply(label_positive, tf.log(pred_positive + epsilon)) +
                        tf.multiply((1-label_positive), tf.log(1. - pred_positive + epsilon)) +
                        tf.multiply(label_negative, tf.log(pred_negative + epsilon)) +
                        tf.multiply((1-label_negative), tf.log(1. - pred_negative + epsilon))
                       )
    return el
    #return -(tf.log(pred_positive + epsilon) + tf.log(1. - pred_negative + epsilon))
#     return tf.reduce_mean(
#         tf.reduce_sum(-(tf.log(pred_positive + epsilon) + tf.log(1. - pred_negative + epsilon)), axis=1)
#     )
```

```python
def predictor_model(input_shape):
    """
    Function creating the context-vector-generator model's graph.
    
    Arguments:
    input_shape -- shape of the correct_ & wrong_ inputs, usually (23,)
    """

    # Create all the layers
    dense5 = Dense(5, activation='relu', name="Dense5")
    dense1 = Dense(1, activation='sigmoid', name="Dense1")
    
    # sentence inputs
    input_0 = Input(shape=input_shape, dtype='float32', name="correct_input")   
   
    y_pred = dense1(dense5(input_0))
    
    # Create Model instance which converts sentence_indices into X.
    model = Model(inputs=[input_0], outputs=[y_pred])
    return model

predictor = predictor_model((23,))
predictor.summary()
predictor.load_weights('/Users/surthi/Documents/cs230project/triplet_loss_custom_function_model.h5', by_name=True)
```
