```python
golds = df[df['isGold'] == 1.0]
g_train, g_test = train_test_split(golds, test_size=0.2)
```

```python
trainArticleIds = g_train['articleId'].to_list()
trainMentions = g_train['mentionMatched'].to_list()
testArticleIds = g_test['articleId'].to_list()
testMentions = g_test['mentionMatched'].to_list()
remainingArticleIds = trainArticleIds + testArticleIds
remainingMentions = trainMentions + testMentions
```

```python
trainMask = (df['articleId'].isin(trainArticleIds) & df['mentionMatched'].isin(trainMentions))
train_df = df.loc[trainMask]
testMask = (df['articleId'].isin(testArticleIds) & df['mentionMatched'].isin(testMentions))
test_df = df.loc[testMask]
remainingMask = (~df['articleId'].isin(remainingArticleIds) & ~df['mentionMatched'].isin(remainingMentions))
remaining_df = df.loc[~trainMask & ~testMask]
remaining_train_df, remaining_test_df = train_test_split(remaining_df, test_size=0.2)
train = pd.concat([train_df, remaining_train_df])
test = pd.concat([test_df, remaining_test_df])
```

```python
concat_df['y_pred_max'] = concat_df.groupby(['articleId', 'mentionMatched'])['y_pred'].transform(max)
```

```python
def f(row):
    if row['y_pred_max'] == row['y_pred']:
        val = row['y_pred_max']
    else:
        val = 0.0
    return val

concat_df['isGold_pred'] = concat_df.apply(f, axis=1)
```
