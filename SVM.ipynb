```python
import pandas as pd
import numpy as np
import sklearn as sns
import numpy as np, pandas as pd, matplotlib.pyplot as plt
from sklearn import ensemble, metrics, model_selection, preprocessing
from sklearn.model_selection import GridSearchCV
from sklearn import preprocessing
from sklearn.model_selection import StratifiedShuffleSplit

# Import function
from sklearn.model_selection import train_test_split

# select features
df = pd.read_json('/Users/surthi/Downloads/Archive (1)/mixed_features_gold_tagged_cleared_betterN/part-00000-e97dbd07-b5c8-4160-9c31-6430e44a03ac.json',lines=True)

df.head()

```

```python
from sklearn import svm
from sklearn.metrics import confusion_matrix
from sklearn import tree, metrics, model_selection, preprocessing

svr = svm.SVC(max_iter=10000,random_state=0)

# #############################################################################
# Train classifiers
#
# For an initial search, a logarithmic grid with basis
# 10 is often helpful. Using a basis of 2, a finer
# tuning can be achieved but at a much higher cost.

C_range = np.logspace(-2, 10, 13)
gamma_range = np.logspace(-9, 3, 13)
param_grid = dict( C=C_range, gamma=gamma_range)
cv = StratifiedShuffleSplit(n_splits=5, test_size=0.2, random_state=0)
grid = GridSearchCV(svr, param_grid=param_grid, cv=cv)

print(param_grid)
print(grid)
```

```python
from sklearn.preprocessing import StandardScaler


# df['species_label'], _ = pd.factorize(df['iris'])
y = df['isGold']
X = df[['w2vScoreN','personScore','sumOfIndustryN',
        'countOfOccurenceOfAliasesN','countOfSignificantStringsN',
        'countryMatchByOverlapScore','directCityMatchScore','directCountryMatchScore',
        'directStateMatchScore','exactMatch','isIndustry','isProminentByItself','articleLengthN',
        'isProminentByParent','isSingleCompany','isTickerMatch','isUrlMatched','stateMatchByOverlapScore'
       ]]


scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# split data randomly into 70% training and 30% test
X_train, X_test, y_train, y_test = model_selection.train_test_split(X_scaled, y, test_size=0.3, random_state=0)


grid.fit(X_train, y_train)
predicted_scaled = grid.predict(X_test)
cnf_matrix_scale = confusion_matrix(y_test, predicted_scaled)
print(cnf_matrix_scale)
```

```python
print('HI')

grid.estimator.get_params()
```

```python

count_misclassified = (y_test != predicted_scaled).sum()
print('Misclassified samples: {}'.format(count_misclassified))

accuracy = metrics.accuracy_score(y_test, predicted_scaled)
print('Accuracy: {:.2f}'.format(accuracy))

SVM_test=df.copy()
SVM_test['y_test']=y_test
SVM_test['y_pred']=pd.DataFrame(predicted_scaled)

from sklearn.metrics import confusion_matrix
from sklearn.metrics import recall_score
from sklearn.metrics import precision_score
confusion_matrix(y_test, predicted_scaled)

print (recall_score(y_test, predicted_scaled))
print(precision_score(y_test, predicted_scaled))

SVM_test.to_json('/Users/surthi/Downloads/svm_scale_predict.json',orient='records', lines=True)
```
